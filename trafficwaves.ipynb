####**Introdução:**
Da mecânica de fluidos, podemos escrever uma equação de movimento tal que: 
<br> 
$$\frac{\partial \rho}{\partial t} = - \nabla \cdot Q(\rho) \tag{1}$$
<center>
tal que $\rho$ é uma densidade (massa, energia ou momentum) e $Q(\rho)$ é um fluxo (massa, energia ou momentum).
</center>
<br>
Considerando apenas uma dimensão ($x$) temos que (1) torna-se: 
<br>
$$\frac{\partial \rho}{\partial t} = - \frac{\partial}{\partial x}Q(\rho) \tag{2}$$

A equação (2) acima, quando no contexto de densidade $\rho$ e fluxo $Q$ de massa, recai sobre a já conhecida equação da continuidade, na qual $\rho$ é uma função da posição ($x$) e do tempo ($t$): $\rho(x,t)$; o fluxo de massa pode ser dado em função de $\rho$ e $v$ tal que: $Q=\rho v$. Onde $v$ é a velocidade associada a massa e pode ser escrita em função de $x$ e $t$: $v(x,t)$. Temos então que (2) torna-se:
<br>
$$\frac{\partial}{\partial t}\rho(x,t) = - \frac{\partial}{\partial x}\bigg[ \rho(x,t) v(x,t)\bigg] \tag{3}$$

####**Trânsito de Veículos:**

Vamos aproximar o trânsito de veículos em uma faixa única e sem ultrapassagens, define-se então as seguintes propriedades desta faixa:
<center>
$\rho(x,t) \equiv $   densidade (carros/km) $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(4.1)$  <br>
$v(\rho) \equiv $   velocidade dos carros (km/hora) $\;\;\;\;\;\;\;\;\;(4.2)$<br>
$q(x,t)  \equiv$ fluxo de carros (carros/hora) $\;\;\;\;\;\;\;\;\;\;\;(4.3)$
</center>
Nota-se que, diferentemente de (3), na equação (4.2), a velocidade depende apenas de $\rho$ e então, da equação (4.3) podemos escrever a relação entre o fluxo de carros, a densidade e a velocidade de uma região desta faixa na forma de:
$$q=\rho v \tag{4.5}$$

Ainda nesta mesma faixa $x$, sabe-se que a quantidade de carros ($n$) dentro de um intervalo $[x, x+\Delta x]$ é dada pela integral da função densidade $\rho$ tal que: 
<br>
 $$ \int_{x}^{x+ \Delta x} \rho(x,t) dx = n \tag{5}$$


image.png

Porém, quando analisamos  a taxa temporal com a qual (5) varia no tempo, faz-se possível a seguinte dedução:
$$ \partial t \int_{x}^{x+ \Delta x} \rho(x,t) dx = q(x)-q(x+\Delta x)\tag{6} $$
Uma vez que (6) se dá ao fato de que a quantidade de carros no intervalo $[x, x+\Delta x]$ vai depender apenas do fluxo de carros que está "entrando" em $x$  e do fluxo de carros que estão "saindo" em $x+\Delta x$. Pode-se fazer a comparação com a função de conservação de massa já conhecida, uma vez que estamos considerando que não existem carros surgindo e sendo destruídos, a quantidade de massa (carros) total em um determinado $t_0$, deve ser a mesma  que em outro $t_0+\Delta t$.

Tomando o limite com $\Delta x \to 0$ em (6), temos:
$$ lim_{\Delta x \to 0} \; \partial t \int_{x}^{x+ \Delta x} \rho(x,t) dx = q(x)-q(x+\Delta x) \Longrightarrow \partial t [\Delta x \rho(x,t)]=\frac{q(x)-q(x + \Delta x)}{\Delta x} \tag{7}$$

Segue-se então,
$$ \partial t [\Delta x \rho(x,t)]=\frac{q(x)-q(x + \Delta x)}{\Delta x} \Longrightarrow \frac{\partial \rho}{\partial t} = -\frac{\partial q}{\partial t}  \tag{8}$$
Nota-se então que, como $q = \rho(x,t)v(\rho)$ uma vez que estmamos considerando que a velocidade $v$ depende apenas da densidade, podemos escrever, a partir de (8) a seguinte equação:
$$ \frac{\partial \rho}{\partial t} = -\frac{\partial q}{\partial t}  \tag{9} \Longrightarrow \frac{\partial }{\partial t}\rho(x,t) = - \frac{\partial }{\partial x} \rho(x,t)v(\rho)$$
Temos assim que, apesar da aproximação na função da velocidade citada anteriormente, (9) é uma equação da continuidade assim como (3) porém para a quantidade de carros e não de massa. Com isso, validamos nossa modelagem do trânsito de veículos.

####**Análise do Problema Não-Linear**
Como $v$ é uma função de $\rho$, com a regra da cadeia, pode-se reescrever a equação (9) de maneira que seja possível tratar o problema apenas em função de $\rho$, já que não temos ainda uma expressão para a função $v(\rho)$; segue-se então:
$$ \frac{\partial}{\partial t}\rho + \frac{\partial}{\partial x}v(\rho)\rho = 0 \Longrightarrow  \frac{\partial}{\partial t}\rho + \frac{\partial \rho}{\partial x}\bigg[v'(\rho)\rho\;+\;v(\rho)\bigg] \therefore \frac{\partial \rho}{\partial t} + c(\rho)\frac{\partial \rho}{\partial x} = 0 \tag{10}$$
Com $c(\rho) = v'(\rho)\rho\;+\;v(\rho)$ para simplificarmos a notação.

Para conseguirmos analisar este problema precisamos definir o que vamos considerar de $v(\rho)$. Neste problema tratado aqui, vamos fazer as seguintes imposições:
<br>
> Com $\rho = 0 \longrightarrow v(\rho) = v_o$ (limite de velocidade da faixa);
<br>
> Com $\rho = \rho_{max} \longrightarrow v(\rho) = 0$ (Congestionamento total da faixa.

Considerando uma relação linear entre $v$ e $\rho$ a partir das imposíções acima,temos a seguinte relação:
$$ v = \frac{v_o}{\rho_{max}}\big(\rho_{max}-\rho\big) \tag{11}$$

```python
#@title 
import matplotlib.pyplot as plt
import numpy as np
import math

rho_max = 1
rho_min = 0
vo = 1

def v(rho):
  return (vo/rho_max)*(rho_max-rho)

fig, ax = plt.subplots(figsize=(16,9))

ax.tick_params(axis="x", labelsize=20)
ax.tick_params(axis="y", labelsize=20)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['right'].set_linewidth(0)
ax.spines['top'].set_linewidth(0)

rho = np.linspace(rho_min, rho_max, 100)
plt.xticks(ticks = [0,1], labels=[r'$\rho_{min}$', r'$\rho_{max}$'])
plt.yticks(ticks = [0,1], labels=[0, r'$v_o$'])
plt.title(r'Gráfico da funçao $v(\rho)$', fontsize=30)
plt.ylabel(r'$v$', fontsize=20, rotation=0)
plt.xlabel(r'$\rho$', fontsize=20)
plt.plot(rho, v(rho), lw=3, label=r'$v(\rho)$')
plt.legend(fontsize=20)



plt.show()
```

<center> Fig. 3 - Gráfico da função $v(\rho)$ a partir da expressão 11. </center>

Partindo de (11), podemos construir uma equação para $q$ tal que:
$$ q = \rho v = \rho \bigg[\frac{v_o}{\rho_{max}}\big(\rho_{max}-\rho\big)\bigg] \tag{12}$$


```python
#@title 
rho_max = 1
rho_min = 0
vo = 1

def q(rho):
  return 2*(vo/rho_max)*rho*(rho_max-rho)

fig, ax = plt.subplots(figsize=(16,9))

ax.tick_params(axis="x", labelsize=20)
ax.tick_params(axis="y", labelsize=20)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['right'].set_linewidth(0)
ax.spines['top'].set_linewidth(0)

rho = np.linspace(rho_min, rho_max, 100)
plt.xticks(ticks = [0,1/2,1], labels=[r'$\rho_{min}$',r'$\frac{1}{2}\rho_{max}$', r'$\rho_{max}$'])
plt.yticks(ticks = [0,1/2], labels=[0, r'$\frac{1}{2}v_o$'])
plt.title(r'Gráfico da funçao $q$', fontsize=30)
plt.ylabel(r'$v$', fontsize=20, rotation=0)
plt.xlabel(r'$\rho$', fontsize=20)
plt.plot(rho, q(rho), lw=3, label=r'$q(v_o,\rho)$', c='green')
plt.ylim(0,1/2+0.1)
plt.legend(fontsize=20)



plt.show()
```

<center> Fig. 4 - Gráfico da função do fluxo. Pode-se perceber que o fluxo tem seu máximo quando $v=\frac{v_o}{2}$ e $\rho = \frac{rho_{max}}{2}$.</center>

Precisamos analisar o que acontece com o termo $c(\rho)$ a partir das definições de (11) e (12).

```python
#@title 
rho_max = 1
rho_min = 0
vo = 1

def c(rho):
  return (vo/rho_max)*(rho_max-2*rho)

fig, ax = plt.subplots(figsize=(16,9))

ax.tick_params(axis="x", labelsize=20)
ax.tick_params(axis="y", labelsize=20)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['right'].set_linewidth(0)
ax.spines['top'].set_linewidth(0)

rho = np.linspace(rho_min, rho_max, 100)
plt.xticks(ticks = [0,1/2,1], labels=[r'$\rho_{min}$',r'$\frac{1}{2}\rho_{max}$', r'$\rho_{max}$'])
plt.yticks(ticks = [-1,0,1], labels=[r'$-v_o$',0, r'$v_o$'])
plt.title(r'Gráfico da funçao $c$', fontsize=30)
plt.ylabel(r'$v$', fontsize=20, rotation=0)
plt.xlabel(r'$\rho$', fontsize=20)
plt.plot(rho, c(rho), lw=3, label=r'$c(v_o,\rho)$', c='orange')
#plt.ylim(0,1/2+0.1)
plt.legend(fontsize=20)



plt.show()
```

####**Solução Analítica - Problema do Valor Inicial**

Desejamos resolver o seguinte problema em questão:

$$\frac{\partial \rho}{\partial t} + c(\rho)\frac{\partial \rho}{\partial x} = 0$$

Sendo as condições iniciais do problema tais que:

$$\rho (x,0) = \rho _0 (x)$$

Com foco nas curvas caracteristicas, se desejarmos $\rho (X(t),t) = const.$ então:

$$\frac{d}{dt}\rho (X(t),t) = \frac{dX}{dt}\frac{\partial \rho}{\partial x} + \frac{\partial \rho}{\partial t} = 0$$

Pelas definições já feitas, obtemos que:

$$\frac{dX}{dt} = c(\rho)$$

As linhas caracteristicas são portanto retas com inclinação dada por $c(\rho _0 (x_0))$, a equação geral é portanto:

$$X(t) = x_0 + c(\rho _0 (x_0))t$$

Não podemos escrever uma solução para o problema inicial mas é possível constatar:

$$\rho (x_0 + c(\rho _0 (x_0))t, t) = \rho _0 (x_0)$$

É possível portanto calcular a solução adiante do tempo dado as condições iniciais do problema mas não o contrário (tendo apenas x e t) pois não temos informações sobre $c(\rho _0 (x_0))$ salvo sabendo as condições iniciais do problema.

```python
def disposition(x):
  return 0.5*(1-math.tanh(x))

x_space = np.linspace(-5,5,200)
disposition_x = np.vectorize(disposition)
po = disposition_x(x_space)
```

####**Resolução via FTCS**


Partimos nossa análise da equação da continuidade:
$$\frac{\partial}{\partial t}\rho (x,t) = -\frac{\partial}{\partial x}Q
(\rho)$$

$$F(\rho) = \rho (x,t)v(\rho (x,t))$$

$$v(\rho) = v_m(1 - \rho/\rho_{max})$$

O método LTCS nos dá a solução tal que:

$$\rho^{n+1}_{i} = \rho^{n}_i - \frac{\tau}{2h}(F^{n}_{i+1} - F^{n}_{i-1})$$

Para o caso initial:

$$\rho _0 = \frac{1 - tanh(x)}{2}$$

```python
def initial_p(x):
  return 0.5*(1-math.tanh(x))

def f_p(p):
  return p*(1-p)

x_space = np.linspace(-50,100,200)
t_space = np.linspace(0,50,51)
acumulate_values = np.zeros((t_space.shape[0],x_space.shape[0]))

set_initial_p = np.vectorize(initial_p)

po = set_initial_p(x_space)

fig, ax = plt.subplots(figsize=(16,9))

ax.tick_params(axis="x", labelsize=20)
ax.tick_params(axis="y", labelsize=20)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)

for time in range(0,t_space.shape[0]):
  p = po
  for position in range(0,p.shape[0]-1):
    p[position] = po[position] - 0.5*(f_p(p[position+1]) - f_p(p[position-1]))
  po = p
  acumulate_values[time] = p
  plt.plot(x_space,p, lw=3)
plt.xlabel(r"$x(t)$", fontsize=20)
plt.ylabel(r"$\rho(x,t)$", fontsize=20)
plt.show()
```
